```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Load the Breast Cancer Dataser from scikit-learn
from sklearn.datasets import load_breast_cancer

# Create instance of the data set
cancer_data = load_breast_cancer()

# This variable (cancer_data) holds more than just the breast cancer data set
raw_data = pd.DataFrame(cancer_data['data'], columns = cancer_data['feature_names'])

print(raw_data)
```

```python
print(cancer_data['DESCR'])
```

```python
x = raw_data
y = cancer_data['target']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
# SVM training
from sklearn.svm import SVC
model = SVC()
model.fit(x_train, y_train)
```

```python
# Making Preedictions
predictions = model.predict(x_test)
```

```python
# Assessing the performance of our support vector machines model
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```
