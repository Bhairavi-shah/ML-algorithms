```python
# Import make_blobs function to create artificial data
from sklearn.datasets import make_blobs

# Use make_blobs function to create artificial data!
# Create data set with 200 samples that has 2 features and 4 cluster centers. The standard deviation within each cluster will be set to 1.8
raw_data = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std = 1.8)

# This is actually a Python tuple. First element of this tuple is a NumPy array with 200 observations. Each observation contains 2 features
print(raw_data)
```

```python
import pandas as pd
import numpy as np
import seaborn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Visualizing Our Data Set

# We see exactly 4 clusters just like we specified
plt.scatter(raw_data[0][:,0], raw_data[0][:,1], c = raw_data[1])
```

```python
# Building and training our K Means clustering model
from sklearn.cluster import KMeans

# Create instance with n_clusters = 4
model = KMeans(n_clusters = 4)

# Fit data
model.fit(raw_data[0])
```

```python
# Make predictions

# 1. Predict which cluster each data point belongs to
print(model.labels_)
```

```python
# 2. Where center of each cluster lies
print(model.cluster_centers_)
```

```python
# Visualizing the accuracy of our model
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))

ax1.set_title('Our Model')

ax1.scatter(raw_data[0][:,0], raw_data[0][:,1],c=model.labels_)

ax2.set_title('Original Data')

ax2.scatter(raw_data[0][:,0], raw_data[0][:,1],c=raw_data[1])
```
