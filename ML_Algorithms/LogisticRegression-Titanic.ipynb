```python
# Removing Null Data From Our Data Set
# We first examine where our data set contains missing data
titanic_data.isnull()
```

```python
# Another way to create a quick visualization is by using a Heatmap
sns.heatmap(titanic_data.isnull(), cbar=False)
```

The white lines indicate the missing values in the dataset.
As we can see, Age & Cabin columns contains the missing data in the Titanic data set

Age column contains a small amount of missing data, so we can fill in the missing data with some form of mathematics.

Cabin data is missing enough data that we could probably remove it from our model entirely

The process of filling in missing data is called **imputation**.

```python
# We will fill in the missing Age values with the average Age value for the specific Pclass passenger class that the passenger belongs to.
sns.boxplot(titanic_data['Pclass'], titanic_data['Age'])
```

The passengers with a Pclass values of 1 (the most expensive class) tend to be the oldest while the passengers with a Pclass value of 3 (the cheapest) tend to be the youngest.

Hence, we use average Age value with different Pclass data to imputate the missing data in our Age column.

```python
def impute_missing_age(columns):
    age = columns[0]
    passenger_class = columns[1]

    if pd.isnull(age):
        #Pclass 1
        if(passenger_class == 1):
            return titanic_data[titanic_data['Pclass'] == 1]['Age'].mean()

        #Pclass 2
        elif(passenger_class == 2):
            return titanic_data[titanic_data['Pclass'] == 2]['Age'].mean()
        
        #Pclass 3
        elif(passenger_class == 3):

            return titanic_data[titanic_data['Pclass'] == 3]['Age'].mean()

        

    else:

        return age
```

```python
# Apply imputation function to every row in the dataframe
titanic_data['Age'] = titanic_data[['Age', 'Pclass']].apply(impute_missing_age, axis = 1)
```

```python
# Confirming imputation
sns.heatmap(titanic_data.isnull(), cbar=False)
```

```python
# Building a Logistic Regression Model

# Remove columns with too much missing data ie, Cabin column
titanic_data.drop('Cabin', axis=1, inplace=True)

# Drop columns that contain missing data
titanic_data.dropna(inplace = True)
```

```python
# Handling categorical data with dummy variables
sex_data = pd.get_dummies(titanic_data['Sex'], drop_first = True)

embarked_data = pd.get_dummies(titanic_data['Embarked'], drop_first = True)
```

```python
# Adding dummy variables to the pandas DataFrame
# We need to add our sex and embarked columns to the DataFrame
titanic_data = pd.concat([titanic_data, sex_data, embarked_data], axis = 1)

titanic_data.columns
```

```python
# Removing Unnecessary Columns From The Data Set
# ie, we can drop the original Sex and Embarked columns from the DataFrame
# We also remove Name, PassengerId, Ticket since they are not predictive of Titanic crash survival rates
titanic_data.drop(['PassengerId', 'Name', 'Ticket', 'Sex', 'Embarked'], axis = 1, inplace = True)
titanic_data.columns
```

```python
titanic_data.head()
```

Every field in the training data and test data is now numeric, which makes it an excellent candidate for a logistic regression machine learning algorithm

```python
# Create training data and test data
y_data = titanic_data['Survived']
x_data = titanic_data.drop('Survived', axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.3)
```

```python
# Training the Logistic Regression Model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
# Make predictions
predictions = model.predict(x_test)
```

```python
# Measuring the Performance of a Logistic Regression Machine Learning Model
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
# Generate a confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, predictions))
```
